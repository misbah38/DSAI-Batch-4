```python
import numpy as np
from sklearn.cluster import KMeans
import cv2
```

```python
testing_link =  r'E:\DSAI batch 4\Machine Learning\HAAR\haar-14Oct2014\temp\positive\rawdata\\' # testing link generate kia hai. end mai zarorat hogi
with open(r'E:\DSAI batch 4\Machine Learning\HAAR\haar-14Oct2014\temp\positive\k_means.txt',"r") as d:   # apni marking wali file open ki aur us ka naam rakh dia 'd'
    raw_data = d.read()   # data read kerwa k usay 1 variable mai daal dia
```

```python
print (raw_data) # check ker lia data ko
```

```python
data = raw_data.split('\n') # ye har record ko new line k through separate ker dia
data
```

```python
dictt = {}                # khali dictionary banayi
for x in data:             # loop chalaya
    y=x.split(" ",1)       # split kia hai record ko 2 tukron mai with each space (" "). agay 1 ka matlab hai k 1st space ko le k 2 tukray banaye ga 
    #print(y)
    dictt[y[0]] = [x.split(" ")[2:] [i:i+4] for i in range(0,len(x.split(" ")[2:]), 4)] # ye mjhaen 
    # print(x.split(" "))
    # PHir record ko split kia spaces se . 
dictt
#and then loop laga dia . ji .sahi. hamain index 2 se onwards ka data chaiye is liye 2: lagaya. 
```

```python
dat =[]   # blank list

for dict_key in dictt.keys():    # yahan se hamain file name milen ge     
	for object_ in dictt[dict_key]:         #values uthayi hain
		dat.append(object_)
        

X = np.array(dat) # yahan numpy array banaya hai values ka
X_ = []   # blank list generated
for x_ in X:    # 
    print(int(x_[2])/int(x_[3]))     # x'204'[0], y'212'[1], width'36'[2], height'43'[3]. width / height kia hai. is ki base p clustering kerwayen ge kmeans ses
    X_.append([int(x_[2])/int(x_[3])])
kmeans = KMeans(n_clusters = 3,n_init=10)  # 2 clusters banaye kmean k.ok
kmeans.fit(X_)    # model banaliya 
labels=kmeans.labels_ # ye agya smjh
labels
centre=kmeans.cluster_centers_
cars = X[labels == 1]
bikes=X[labels==2]
humans = X[labels == 0]
humans
```

```python
def bounding(image_path, arr, kind):  # function banaya

    image = cv2.imread(image_path)     # variable banaya image read kerne k liye. 6 baj gaye han kal cont krskte ?ji jithankyou

    if image is None:
        print(f"Error: Could not read image at {image_path}")
        return
        
    x, y, width, height = int(arr[0]), int(arr[1]), int(arr[2]), int(arr[3])
    
    if kind == "humans":
        color = (0, 255, 255)
        thickness = 2
    if kind== "bikes":
        color = (255, 255, 0)
        thickness = 2
    if kind == "cars":
        color = (255, 0,255)
        thickness = 2
        
    cv2.rectangle(image, (x, y), (x + width, y + height), color, thickness)

    cv2.imshow('Image with Bounding Box', image)
    cv2.waitKey(0)  # Wait for a key press before closing the window
    cv2.destroyAllWindows()
```
