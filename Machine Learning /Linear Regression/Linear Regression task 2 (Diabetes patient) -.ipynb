```python
# now set the values for independent and dependent in to features and label variable (X, y)
features= ['Glucose','BloodPressure','Insulin','BMI','Age']
X= df[features]
X=X.iloc[:,:-1].values
X
print(len(X))
```

```python
# those have value above form 0.5 are diabetic patient
y = df[['DiabetesPedigreeFunction']]
y=y.iloc[::1].values
y
print(len(y))
```

```python
#now split the data in to Taining and testing data 
# 0.3 shows that 30% data is for testing whereas 70% data is for training
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

#1937*9 rows
X_train.shape

#1937*1
y_train.shape #(total training transactions)

#831*9 rows
X_test.shape

#831, 1
y_test.shape #(total testing Transactions)
```

### Linear Regression
*  Linear regression analysis is used to predict the value of a variable based on the value of another variable. The variable you want to predict is called the dependent variable. The variable you are using to predict the other variable's value is called the independent variable.




```python

from sklearn.linear_model import LinearRegression
# Create a Linear Regression model
model = LinearRegression()
```

```python
# Train the model on the training data
model.fit(X_train, y_train)

```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(len(y_pred))
```

```python
from sklearn.metrics import r2_score

rsquare = r2_score(y_test, y_pred)
print("Mean Squared:", rsquare)
```

```python
from sklearn.metrics import mean_squared_error
# Calculate the Mean Squared Error (MSE) as a measure of model performance
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Print the coefficients and intercept of the linear regression model
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_test, y_pred, color='red',linewidth=3, label='Linear Regression')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression')
plt.legend()
plt.show()
# Print the coefficients (slope and intercept)
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# Visualize the model's predictions vs. actual values
plt.scatter(y_test, y_pred)
plt.xlabel("Atual Diabetic Probability")
plt.ylabel("PredictedDiabetic Probability")
plt.title("Actual vs. Predicted Diabetic Probability")
plt.show()
```

```python
# Plot the graph after applying Linear Regression
plt.scatter(X, y, label='Original Data')
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression')
plt.legend()
plt.show()
# Print the coefficients (slope and intercept)
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
